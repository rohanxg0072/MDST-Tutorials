<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
# setting x and y

X = df[['INSTRUCTION_EXPENDITURE','CAPITAL_OUTLAY_EXPENDITURE','SUPPORT_SERVICES_EXPENDITURE']].dropna()
y = df.loc[X.index]['AVG_MATH_8_SCORE']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=.4, random_state=0)
```

<h2> Prediction </h2>

ML Models Resource: https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a

Chosen ML task: REGRESSION

```python
# import your sklearn class here

from sklearn.linear_model import LinearRegression
```

```python
# create your model here
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# FOR CLASSIFICATION ONLY:
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(model, X_test, y_test,
                         cmap=plt.cm.Blues)
```

```python
# FOR REGRESSION ONLY: (pick a single column to visualize results)

# Results from this graph _should not_ be used as a part of your results -- it is just here to help with intuition. 
# Instead, look at the error values and individual intercepts.


col_name = "INSTRUCTION_EXPENDITURE"
col_index = X_train.columns.get_loc(col_name)

f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "red")
plt.scatter(X_train[col_name], model.predict(X_train), color = "green")
plt.scatter(X_test[col_name], model.predict(X_test), color = "blue")

new_x = np.linspace(X_train[col_name].min(),X_train[col_name].max(),200)
intercept = model.predict([X_train.sort_values(col_name).iloc[0]]) - X_train[col_name].min()*model.coef_[col_index]
plt.plot(new_x, intercept+new_x*model.coef_[col_index])

plt.legend(['controlled model','true training','predicted training','predicted testing'])
plt.xlabel(col_name)
plt.ylabel("Average 8th grade math score")

# printing coefficients and error
print(model.coef_)
print(np.mean(np.abs(model.predict(X_test)-y_test)))
```

<h2> Summary </h2>

I set out to use machine learning to determine the impact of instruction expenditure, capital outlay expenditure and support services expenditure on average eighth grade math scores. I analyzed absolute spending in each category instead of spending per capita, for absolute spending represents a combination of the size of a state and the amount per student it funds education; I wonderered if any interesting patterns might emerge when looking at the data this way. I chose a regression model because I wanted to see quantitatively how each of my variables impacted test scores, and I specifically chose linear regression because it was a model I was somewhat familiar with from everyday life.

Based on my coefficients, it seems as though absolute instruction and support spending correlate positively with test scores, while capital outlay spending correlates negatively. Yet, all of these coefficients were quite small, which, along with the large mean error of around 8, leads me to question whether the variables I chose had any meaningful impact on test scores. Ultimately, my methodology of using absolute rather than per-student funding levels was likely flawed and unhelpful, leading to the low coefficients and high error in my model, which means we likely learned little from this analysis and model.
